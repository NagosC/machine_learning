```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.compose import make_column_transformer
```

```python
dataset = pd.read_csv(r'C:\Users\leand\OneDrive\Área de Trabalho\CursoCódigos\34.Prática em Python\dados\Credit2.csv', sep = ';')
dataset.shape
```

```python
#separação das variáveis, ingnoro a priemeira pois não tem valor semantico 
x = dataset.iloc[:,1:10].values
y = dataset.iloc[:,10].values

x
```

```python
#label encoder para a coluna cheking_status
#Atribui os valores de zero a 3
labelEncoder = LabelEncoder()
x[:,0] = labelEncoder.fit_transform(x[:,0])

x
```

```python
#onehotencode para a coluna credit history 
hotencoder = make_column_transformer((OneHotEncoder(categories = 'auto', sparse = False), [1]), remainder = 'passthrough')
x = hotencoder.fit_transform(x)
x
```

```python
#excluimos a variável para evitar a dummy variable trap
x = x[:,1:]
```

```python
#label enconder com a classe 
labelencoder_y = LabelEncoder()
y = labelencoder_y.fit_transform(y)
y
```

```python
#processo de separação treino e teste
X_treino, X_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2,random_state = 0)
```

```python
#feature Scalling,  Padronização z-score
sc = StandardScaler()
X_treino = sc.fit_transform(X_treino)
X_teste = sc.fit_transform(X_teste)
```

```python
classifier = Sequential()
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 12))
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
classifier.fit(X_treino, y_treino, batch_size = 10 , epochs = 100  )
```

```python
y_pred = classifier.predict(X_teste)
y_pred = (y_pred > 0.5)
y_pred
```

```python
#matriz de confusão
cm = confusion_matrix(y_teste, y_pred)
cm
```
